```python
import librosa
import numpy as np
import os
import pandas as pd
import glob

input_dir = "extracted_audios"
pattern = os.path.join(input_dir, "**", "*.wav")
wav_files = glob.glob(pattern, recursive=True)
print(f"Found {len(wav_files)} .wav files")

n_mfcc = 40  # Increased number of MFCC coefficients

data = []

for path in wav_files:
    try:
        y, sr = librosa.load(path, sr=None)
        if y.size == 0:
            print(f"Empty file: {path}")
            continue
        mfcc = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=n_mfcc)
        mfcc_mean = mfcc.mean(axis=1)
        filename = os.path.basename(path)
        data.append([filename] + mfcc_mean.tolist())
    except Exception as e:
        print(f"Error processing {path}: {e}")

if data:
    cols = ["filename"] + [f"mfcc{i}" for i in range(1, n_mfcc + 1)]
    df = pd.DataFrame(data, columns=cols)
    df.to_csv("mfcc_features.csv", index=False)
    print("MFCCs saved to mfcc_features.csv")
else:
    print("No valid MFCC data found.")
```
