```python
import os
import glob
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np
import zipfile

# Define the input zip file, temporary extraction directory, and output directory for spectrograms
zip_file = "Trimmed Audios.zip"
extract_dir = "extracted_audios"
output_dir = "spectrograms"
os.makedirs(extract_dir, exist_ok=True)
os.makedirs(output_dir, exist_ok=True)

# Step 1: Unzip the Trimmed Audios.zip file
try:
    with zipfile.ZipFile(zip_file, 'r') as zip_ref:
        zip_ref.extractall(extract_dir)
    print(f"Extracted {zip_file} to {extract_dir}")
except FileNotFoundError:
    print(f"Error: The file {zip_file} was not found in the current directory.")
    exit(1)
except zipfile.BadZipFile:
    print(f"Error: {zip_file} is not a valid zip file.")
    exit(1)
except Exception as e:
    print(f"Error during unzipping: {str(e)}")
    exit(1)

# Step 2: Get list of all .wav files recursively from the extracted directory
wav_files = glob.glob(os.path.join(extract_dir, "**", "*.wav"), recursive=True)
print(f"Found {len(wav_files)} .wav files in {extract_dir}")

# Parameters for spectrogram
image_shape = (224, 448)  # Increased width for longer x-axis
n_fft = 2048
hop_length = 512
n_mels = 128

# Step 3: Process each .wav file to generate spectrograms
for wav_file in wav_files:
    print(f"Processing {os.path.basename(wav_file)}...")
    try:
        # Load audio file
        y, sr = librosa.load(wav_file, sr=44100)
        print(f"Loaded audio: {os.path.basename(wav_file)}, sample rate: {sr}, samples: {len(y)}")

        # Compute mel spectrogram
        S = librosa.feature.melspectrogram(y=y, sr=sr, n_fft=n_fft, hop_length=hop_length, n_mels=n_mels)
        S_dB = librosa.power_to_db(S, ref=np.max)

        # Create a new figure with increased width for longer x-axis
        plt.figure(figsize=(image_shape[1]/50, image_shape[0]/100), dpi=100)  # Doubled width for x-axis

        # Plot the spectrogram (creates a mappable object)
        mappable = librosa.display.specshow(S_dB, sr=sr, hop_length=hop_length, x_axis='time', y_axis='mel', cmap='viridis')

        # Add colorbar, referencing the mappable object
        plt.colorbar(mappable=mappable, format='%+2.0f dB')

        # Set title and labels
        #plt.title(f"Spectrogram of {os.path.basename(wav_file)}")
        #plt.xlabel("Time")
        #plt.ylabel("Frequency")

        # Create a safe output filename by replacing invalid characters
        base_name = os.path.splitext(os.path.basename(wav_file))[0]
        output_filename = f"{base_name}_spectrogram.png"
        output_path = os.path.join(output_dir, output_filename)

        # Ensure the output filename is unique (handle duplicates)
        counter = 1
        while os.path.exists(output_path):
            output_filename = f"{base_name}_spectrogram_{counter}.png"
            output_path = os.path.join(output_dir, output_filename)
            counter += 1

        # Save the plot to the output directory
        plt.savefig(output_path, bbox_inches='tight')
        print(f"Saved spectrogram: {output_path}")

        # Close the figure to free memory
        plt.close()

    except Exception as e:
        print(f"Error processing {os.path.basename(wav_file)}: {str(e)}")

print("Processing complete.")
```
