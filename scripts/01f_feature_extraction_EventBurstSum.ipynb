```python
import torch
import torchaudio
from torch.utils import data
import IPython.display as display
import numpy as np
import matplotlib.pyplot as plt
import os
import librosa

audio2spec = torchaudio.transforms.Spectrogram(n_fft=300)
spec2audio = torchaudio.transforms.GriffinLim(n_fft=300)

def audio_to_specDB(x):
    x = audio2spec(x)
    x = torchaudio.functional.amplitude_to_DB(x, multiplier=1, amin=1e-10, db_multiplier=0)
    return x

def specDB_to_audio(x):
    x = torchaudio.functional.DB_to_amplitude(x, ref=10, power=1)
    x = spec2audio(x)
    return x

def analyzer(audio_folder):
    items = sorted(os.listdir(audio_folder))
    print('Audio folder:', audio_folder, 'with', len(items), 'items')
    evals = []
    signals = []
    specs = []
    for item in items:
        path = os.path.join(audio_folder, item)
        meta = torchaudio.info(path)
        w, r = torchaudio.load(path)
        spec = audio_to_specDB(w)

        mat = np.array(spec[0])
        min_val = np.min(mat)
        mat = (mat - min_val)
        signal = np.mean(mat[40:], axis=0)
        signal = np.diff(signal, 1)
        signal = np.abs(signal)

        value = np.sum(signal / len(signal))
        evals.append(value)
        signals.append(signal)
        specs.append(spec)
    return evals, specs, signals

list_eva = []
list_spec = []
list_signal = []
list_label = []

path = 'finaldata/'
genotypes = os.listdir(path)

for gen_name in genotypes:
    p = os.path.join(path, gen_name)
    eva, spec, signal = analyzer(p)
    list_eva.append(eva)
    list_spec.append(spec)
    list_signal.append(signal)
    list_label.append(gen_name)
    

average_evals = [np.mean(eva) for eva in list_eva]

sorted_indices = np.argsort(average_evals)
sorted_labels = [list_label[i] for i in sorted_indices]
sorted_evals = [list_eva[i] for i in sorted_indices]

plt.figure(figsize=(20, 10))
box = plt.boxplot(sorted_evals, labels=sorted_labels, vert=True, patch_artist=True)

plt.xlabel('Genotype')
plt.ylabel('Values')
plt.title('Method1, Summing all Changes in Signal for Agrovision')

# Rotate x-axis labels to be vertical
plt.xticks(rotation=90)

# Add scatter plot and median labels
for i in range(len(sorted_evals)):
    y = sorted_evals[i]
    x = np.random.normal(1 + i, 0.04, size=len(y))
    plt.scatter(x, y, alpha=0.5)
    plt.text(i + 1, np.median(y), f'{np.median(y):.2f}', 
             horizontalalignment='center', verticalalignment='center', 
             fontdict={'fontsize': 10, 'weight': 'bold'})

plt.show()
```
