```python
# Tabnet 2c tdf (corrected with 10-fold CV, 3 repetitions, and history padding)

import pandas as pd
import numpy as np
import random
import torch
import pickle
import matplotlib.pyplot as plt
from sklearn.model_selection import RepeatedStratifiedKFold
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import (
    confusion_matrix, accuracy_score, balanced_accuracy_score, cohen_kappa_score,
    f1_score, precision_score, recall_score, log_loss, roc_auc_score,
    precision_recall_curve, auc
)
from sklearn.utils.class_weight import compute_class_weight
from pytorch_tabnet.tab_model import TabNetClassifier
from pytorch_tabnet.callbacks import Callback
import optuna

# === Reproducibility ===
np.random.seed(123)
random.seed(123)
torch.manual_seed(123)

# === Callback for History ===
class CustomHistory(Callback):
    def __init__(self):
        super().__init__()
        self.history = {'train_accuracy': [], 'test_accuracy': [], 'train_loss': [], 'test_loss': []}
    def set_trainer(self, trainer): 
        self.trainer = trainer
    def on_epoch_end(self, epoch, logs=None):
        logs = logs or {}
        self.history['train_accuracy'].append(logs.get('train_accuracy', logs.get('accuracy', 0.0)))
        self.history['test_accuracy'].append(logs.get('test_accuracy', logs.get('valid_accuracy', 0.0)))
        self.history['train_loss'].append(logs.get('train_logloss', logs.get('loss', 0.0)))
        self.history['test_loss'].append(logs.get('test_logloss', logs.get('valid_logloss', 0.0)))
        print(f"Epoch {epoch + 1} logs: {logs}")

# === Load and Prepare Data ===
df = pd.read_csv('../data/01_training_cv_data.csv')

# Define genotypes and corresponding Paul values
genotypes = ["21-1350", "19.75-4", "20-088", "15-282", "17-147", "20-105", "19-67"]
paul_values = [3, 3, 3, 4, 4, 4, 3]

# Create a mapping dictionary from genotype to Paul value
genotype_to_paul = dict(zip(genotypes, paul_values))

# Update the 'Paul' column based on matches
df.loc[df['Genotype'].isin(genotype_to_paul.keys()), 'Paul'] = \
    df.loc[df['Genotype'].isin(genotype_to_paul.keys()), 'Genotype'].map(genotype_to_paul)

df = df.iloc[:, 1:]

df = df.assign(
    r=lambda x: np.where(x['Paul'].isin([1, 2, 3]), 1,
                         np.where(x['Paul'].isin([4, 5]), 2, np.nan))
).drop(columns=['Paul']).dropna()
df['r'] = pd.Categorical(df['r'], categories=[1, 2], ordered=True).rename_categories(['class1', 'class2'])

features = [col for col in df.columns if col != 'r']
scaler = StandardScaler()
df[features] = scaler.fit_transform(df[features])

X = df[features].to_numpy()
y = df['r'].cat.codes.to_numpy()

# === Define Optuna Objective Function ===
def objective(trial, x_train, y_train, x_val, y_val):
    params = {
        'n_d': trial.suggest_int('n_d', 16, 64),
        'n_a': trial.suggest_int('n_a', 16, 64),
        'n_steps': trial.suggest_int('n_steps', 3, 7),
        'gamma': trial.suggest_float('gamma', 1.0, 2.0),
        'lambda_sparse': trial.suggest_float('lambda_sparse', 1e-4, 1e-2, log=True),
        'n_independent': trial.suggest_int('n_independent', 2, 4),
        'n_shared': trial.suggest_int('n_shared', 2, 4),
        'optimizer_fn': torch.optim.Adam,
        'optimizer_params': {'lr': trial.suggest_float('lr', 1e-3, 5e-2, log=True)},
        'scheduler_params': {'step_size': trial.suggest_int('step_size', 5, 20),
                             'gamma': trial.suggest_float('gamma_lr', 0.7, 0.95)},
        'scheduler_fn': torch.optim.lr_scheduler.StepLR,
        'mask_type': 'sparsemax',
        'seed': 123,
        'verbose': 0
    }

    # Validate class weights
    unique_classes = np.unique(y_train)
    if len(unique_classes) < 2:
        print(f"Warning: Not all classes present in training fold. Classes found: {unique_classes}")
        return 0.0  # Return low score to avoid using this trial

    model = TabNetClassifier(**params)
    class_weights = compute_class_weight('balanced', classes=np.unique(y_train), y=y_train)
    sample_weights = np.array([class_weights[label] for label in y_train])

    model.fit(
        X_train=x_train,
        y_train=y_train,
        eval_set=[(x_val, y_val)],
        eval_name=['test'],
        eval_metric=['accuracy'],
        max_epochs=100,
        patience=20,
        batch_size=64,
        virtual_batch_size=16,
        weights=sample_weights
    )
    return accuracy_score(y_val, model.predict(x_val))

# === Cross-validation Loop ===
rkf = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=123)
cv_data, cv_accuracies, histories = [], [], []
last_model = None
studies = []  # Store Optuna studies for debugging

for fold, (train_idx, test_idx) in enumerate(rkf.split(X, y)):
    x_train, x_test = X[train_idx], X[test_idx]
    y_train, y_test = y[train_idx], y[test_idx]

    try:
        study = optuna.create_study(direction='maximize')
        study.optimize(lambda trial: objective(trial, x_train, y_train, x_test, y_test), n_trials=20)
        studies.append(study)
        best = study.best_params

        # Train with best parameters
        model = TabNetClassifier(
            n_d=best['n_d'], n_a=best['n_a'], n_steps=best['n_steps'],
            gamma=best['gamma'], lambda_sparse=best['lambda_sparse'],
            n_independent=best['n_independent'], n_shared=best['n_shared'],
            optimizer_fn=torch.optim.Adam,
            optimizer_params={'lr': best['lr']},
            scheduler_params={'step_size': best['step_size'], 'gamma': best['gamma_lr']},
            scheduler_fn=torch.optim.lr_scheduler.StepLR,
            mask_type='sparsemax', seed=123, verbose=0
        )

        history = CustomHistory()
        class_weights = compute_class_weight('balanced', classes=np.unique(y_train), y=y_train)
        sample_weights = np.array([class_weights[label] for label in y_train])
        
        model.fit(
            X_train=x_train,
            y_train=y_train,
            eval_set=[(x_train, y_train), (x_test, y_test)],
            eval_name=['train', 'test'],
            eval_metric=['accuracy', 'logloss'],
            max_epochs=100,
            patience=20,
            batch_size=64,
            virtual_batch_size=16,
            weights=sample_weights,
            callbacks=[history]
        )

        preds = model.predict(x_test)
        probs = model.predict_proba(x_test)
        cv_data.append(pd.DataFrame({
            'obs': y_test, 'pred': preds,
            'prob.0': probs[:, 0], 'prob.1': probs[:, 1]
        }))
        acc = accuracy_score(y_test, preds)
        print(f"Fold {fold + 1} Accuracy: {acc:.4f}")
        cv_accuracies.append(acc)
        histories.append(history.history)
        last_model = model

    except (ValueError, RuntimeError) as e:
        print(f"Fold {fold + 1} failed: {e}")
        continue

# === Combine CV Results ===
if not cv_data:
    raise ValueError("No folds were completed successfully. Cannot continue.")

cv_data = pd.concat(cv_data, ignore_index=True)
cv_data['obs'] = pd.Categorical(cv_data['obs'].map({0: 'class1', 1: 'class2'}),
                                categories=['class1', 'class2'])
cv_data['pred'] = pd.Categorical(cv_data['pred'].map({0: 'class1', 1: 'class2'}),
                                 categories=['class1', 'class2'])

# === Plot History ===
def pad_history(h, length):
    return {k: h[k] + [np.nan]*(length-len(h[k])) if len(h[k]) < length else h[k][:length]
            for k in h}

max_epochs = max(len(h['train_accuracy']) for h in histories)
padded_histories = [pad_history(h, max_epochs) for h in histories]

avg_history = {
    key: np.nanmean([h[key] for h in padded_histories], axis=0)  # Use nanmean to handle NaNs
    for key in ['train_accuracy', 'test_accuracy', 'train_loss', 'test_loss']
}

import seaborn as sns

# Set Seaborn style
sns.set_style("darkgrid")  # Options: 'darkgrid', 'whitegrid', 'dark', 'white', 'ticks'

# Plotting code
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(avg_history['train_accuracy'], label='Train Accuracy', color='blue')
plt.plot(avg_history['test_accuracy'], label='Test Accuracy', color='orange')
plt.legend()
plt.title('Accuracy Over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(avg_history['train_loss'], label='Train Loss', color='blue')
plt.plot(avg_history['test_loss'], label='Test Loss', color='orange')
plt.legend()
plt.title('Loss Over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.grid(True)
plt.tight_layout()
plt.savefig('training_history_2c.png')
plt.show()
plt.close()

# Verify shapes
print("Number of features:", len(features))
print("Number of importances:", len(last_model.feature_importances_))

# Plot feature importance
if last_model:
    sns.set_style("darkgrid")
    plt.figure(figsize=(8, 6))
    plt.bar(features, last_model.feature_importances_, color='skyblue')
    plt.xticks(rotation=45, ha='right')
    plt.title('Feature Importance')
    plt.xlabel('Features')
    plt.ylabel('Importance')
    plt.tight_layout()
    plt.savefig('feature_importance_2c.png')
    plt.show()
    plt.close()

# === Compute Metrics ===
cm = confusion_matrix(cv_data['obs'], cv_data['pred'], labels=['class1', 'class2'])
accuracy = accuracy_score(cv_data['obs'], cv_data['pred'])
balanced_acc = balanced_accuracy_score(cv_data['obs'], cv_data['pred'])
detection_rate = np.mean([cm[i, i] / cm[i, :].sum() for i in range(2) if cm[i, :].sum() > 0])
f1 = np.mean([f1_score(cv_data['obs'] == cls, cv_data['pred'] == cls) for cls in ['class1', 'class2']])
kappa = cohen_kappa_score(cv_data['obs'], cv_data['pred'])
prob_matrix = cv_data[['prob.0', 'prob.1']].to_numpy()
y_true = cv_data['obs'].cat.codes
logloss = -np.mean([np.log(prob_matrix[i, y_true[i]]) for i in range(len(y_true)) if prob_matrix[i, y_true[i]] > 0])
npv = np.mean([cm[i, i] / cm[:, i].sum() for i in range(2) if cm[:, i].sum() > 0])
ppv = np.mean([precision_score(cv_data['obs'] == cls, cv_data['pred'] == cls) for cls in ['class1', 'class2']])
sensitivity = recall_score(cv_data['obs'], cv_data['pred'], average='macro')
specificity = np.mean([cm[i, i] / cm[:, i].sum() for i in range(2) if cm[:, i].sum() > 0])
precision = ppv
recall = sensitivity
auc_score = roc_auc_score(y_true, prob_matrix[:, 1])
pr_aucs = []
for cls in [0, 1]:
    labels = (y_true == cls).astype(int)
    scores = prob_matrix[:, cls]
    if len(np.unique(labels)) < 2:
        continue
    precision_vals, recall_vals, _ = precision_recall_curve(labels, scores)
    pr_aucs.append(auc(recall_vals, precision_vals))
pr_auc = np.mean(pr_aucs) if pr_aucs else np.nan

# Create metrics DataFrame
metrics = {
    'Metric': ['Accuracy', 'AUC', 'Balanced Accuracy', 'Detection Rate', 'F1', 'Kappa',
               'LogLoss', 'Neg Pred Value', 'Pos Pred Value', 'prAUC', 'Precision',
               'Recall', 'Sensitivity', 'Specificity'],
    'Value': [
        accuracy, auc_score, balanced_acc, detection_rate, f1, kappa,
        logloss, npv, ppv, pr_auc, precision, recall, sensitivity, specificity
    ]
}
metrics_df = pd.DataFrame(metrics)

print("\nPerformance Metrics:")
print(metrics_df)

# === Save Everything ===
metrics_df.to_excel("tabnet_acoustic_pred_2c.xlsx", index=False)
cv_data.to_csv("tabnet_predictions_2c.csv", index=False)
with open('tabnet_tdf_2c.pkl', 'wb') as f:
    pickle.dump({
        'history': avg_history,
        'caret_fit': cv_accuracies,
        'tabnet_model': last_model,
        'cv_data': cv_data,
        'metrics_df': metrics_df,
        'perm_importance': last_model.feature_importances_ if last_model else None,
        'optuna_studies': studies
    }, f)

print("\nMean CV Accuracy:", np.mean(cv_accuracies))
print("\nPerformance Metrics:")
# print(metrics_df)
```

```python
import pandas as pd
import numpy as np
from sklearn.metrics import (
    confusion_matrix, accuracy_score, balanced_accuracy_score, cohen_kappa_score,
    log_loss, precision_score, recall_score, f1_score, roc_auc_score,
    precision_recall_curve, auc
)

# Load cv_data
cv_data = pd.read_csv('tabnet_predictions_2c.csv')

# Verify probability sums
prob_array = cv_data[['prob.0', 'prob.1']].to_numpy()
prob_sums = prob_array.sum(axis=1)
if not np.allclose(prob_sums, 1, atol=1e-6):
    print("Warning: Probabilities do not sum to 1. Normalizing probabilities.")
    prob_array = prob_array / prob_array.sum(axis=1, keepdims=True)
    prob_array = np.nan_to_num(prob_array, nan=0.5, posinf=0.5, neginf=0.5)
else:
    print("Probabilities are valid (sum to 1).")

# Compute confusion matrix
cm = confusion_matrix(cv_data['obs'], cv_data['pred'], labels=['class1', 'class2'])

# Compute metrics
accuracy = accuracy_score(cv_data['obs'], cv_data['pred'])
misclassification_rate = 1 - accuracy  # Calculate misclassification rate
balanced_acc = balanced_accuracy_score(cv_data['obs'], cv_data['pred'])
kappa = cohen_kappa_score(cv_data['obs'], cv_data['pred'])
logloss = log_loss(cv_data['obs'], prob_array)  # Use normalized probabilities
precision = precision_score(cv_data['obs'], cv_data['pred'], pos_label='class2', average='binary', zero_division=0)
recall = recall_score(cv_data['obs'], cv_data['pred'], pos_label='class2', average='binary', zero_division=0)
f1 = f1_score(cv_data['obs'], cv_data['pred'], pos_label='class2', average='binary', zero_division=0)
sensitivity = recall_score(cv_data['obs'], cv_data['pred'], average='macro', zero_division=0)
specificity = np.mean([cm[i, i] / cm[:, i].sum() for i in range(2) if cm[:, i].sum() > 0])
npv = cm[0, 0] / cm[:, 0].sum() if cm[:, 0].sum() > 0 else 0  # NPV for class1
auc_score = roc_auc_score(cv_data['obs'], cv_data['prob.1'])
precision_vals, recall_vals, _ = precision_recall_curve(cv_data['obs'], cv_data['prob.1'], pos_label='class2')
pr_auc = auc(recall_vals, precision_vals)

# Create metrics DataFrame
metrics = {
    'Metric': ['Accuracy', 'Misclassification Rate', 'Balanced Accuracy', 'Kappa', 'Log Loss', 
               'Precision', 'Recall', 'F1 Score', 'Sensitivity', 'Specificity', 'NPV', 'AUC', 'PR-AUC'],
    'Value': [accuracy, misclassification_rate, balanced_acc, kappa, logloss, precision, recall, f1,
              sensitivity, specificity, npv, auc_score, pr_auc]
}
metrics_df = pd.DataFrame(metrics)

# Display metrics
print("\nPerformance Metrics:")
print(metrics_df)
```
